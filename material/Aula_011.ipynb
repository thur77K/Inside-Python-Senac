Essa consulta com grupo s√≥ pode conter os campos utilizados para compor a chave do grupo e fun√ß√µes de agrupamento de dados, como ```MIN``` (m√≠nimo), ```MAX``` (m√°ximo), ```AVG``` (m√©dia, do ingl√™s *average*), SUM (soma) e ```COUNT``` (contagem).

Um exemplo concreto como nosso banco de dados √© agrupar os estados por regi√£o. A consulta seria algo como:

```sql
SELECT regi√£o, count(*) from group by regi√£o
```

Esse comando utiliza a cl√°usula ```GROUP BY``` regi√£o para especificar a chave de grupo. Dessa forma, todos os registros que pertencem √° mesma regi√£o s√£o agrupados. Observe que os campos ap√≥s o ```SELECT``` incluem regi√£o e ```COUNT(*)```. O campo regi√£o p√¥de ser inclu√≠do, pois faz parte da chave de grupo especificada em ```GROUP BY```. A fun√ß√£o ```COUNT(*)``` retorna quantos registros fazem parte do grupo. Vejamos o resultado do programa:

```python
# T√≠tulo: Exibindo o n√∫mero de estados por regi√£o
##############################################################################
import sqlite3

print("Regi√£o N√∫mero de Estados")
print("====== =================")
with sqlite3.connect("brasil.db") as conex√£o:
    for regi√£o in conex√£o.execute("""
        select regi√£o, count(*)
        from estados
        group by regi√£o"""):
        print("{0:6} {1:17}".format(*regi√£o))
```


Vamos adicionar as fun√ß√µes ```MIN```, ```MAX```, ```SUM``` e ```AVG``` no campo popula√ß√£o. Veja o novo programa:

```python
# T√≠tulo: Exibindo a popula√ß√£o m√≠nima, m√°xima, m√©dia e total de cada regi√£o
##############################################################################
import sqlite3

print("Regi√£o Estados Popula√ß√£o  M√≠nima    M√°xima      M√©dia    Total (soma)")
print("====== =======          ========= ========== ==========  ============")
with sqlite3.connect("brasil.db") as conex√£o:
    for regi√£o in conex√£o.execute("""
        select regi√£o, count(*), min(popula√ß√£o),
               max(popula√ß√£o), avg(popula√ß√£o), sum(popula√ß√£o)
        from estados
        group by regi√£o"""):
        print("{0:6} {1:7} {2:18,} {3:10,} {4:10,.0f} {5:13,}".format(*regi√£o))
    print(
        "\nBrasil: {0:6} {1:18,} {2:10,} {3:10,.0f} {4:13,}".format(
            *conex√£o.execute("""
            select count(*), min(popula√ß√£o), max(popula√ß√£o),
                   avg(popula√ß√£o), sum(popula√ß√£o) from estados""").fetchone()
        )
    )
```


Conseguimos calcular a popula√ß√£o m√≠nima, m√°xima, m√©dia e total de cada regi√£o e tamb√©m para o Brasil. Veja que, na segunda consulta, a que calcula os dados para o Brasil. Veja que, na segunda consulta, a que calcula os dados para o Brasil, n√£o utilizamos a cl√°usula ```GROUP BY```, podemos continuar usando tudo que j√° aprendemos em SQL, como as cl√°usulas ```WHERE``` e ```ORDER BY```. Vejamos o mesmo programa, mas com linhas ordenadas pela popula√ß√£o total de cada regi√£o em ordem decrescente:

```python
# T√≠tulo: Ordenando os resultados
##############################################################################
import sqlite3

print("Regi√£o Estados Popula√ß√£o  M√≠nima    M√°xima      M√©dia    Total (soma)")
print("====== =======          ========= ========== ==========  ============")
with sqlite3.connect("brasil.db") as conex√£o:
    for regi√£o in conex√£o.execute("""
        select regi√£o, count(*), min(popula√ß√£o),
               max(popula√ß√£o), avg(popula√ß√£o), sum(popula√ß√£o)
        from estados
        group by regi√£o
        order by sum(popula√ß√£o) desc"""):
        print("{0:6} {1:7} {2:18,} {3:10,} {4:10,.0f} {5:13,}".format(*regi√£o))
    print(
        "\nBrasil: {0:6} {1:18,} {2:10,} {3:10,.0f} {4:13,}".format(
            *conex√£o.execute("""
            select count(*), min(popula√ß√£o), max(popula√ß√£o),
                   avg(popula√ß√£o), sum(popula√ß√£o) from estados""").fetchone()
        )
    )
```
      

Veja que escrevemos `SUM(popula√ß√£o) as tpop`, dando o nome `tpop` √† soma. Esse tipo de constru√ß√£o evita a repeti√ß√£o da fun√ß√£o na consulta e facilita a leitura.

Podemos tamb√©m filtrar os resultados ap√≥s o agrupamento, usando a cl√°usula `HAVING`. Para entender a diferen√ßa entre `WHERE` e `HAVING`, imagine que `WHERE` √© executada antes antes do agrupamento, selecionando os registros que far√£o parte do resultado, antes de o agrupamento ser realizado. A cl√°usula `HAVING` avalia o resultado do agrupamento e decide quais far√£o parte do resultado final. Por exemplo, podemos escolher apenas regi√µes com mais de cinco estados. Como a quantidade de estados por regi√£o s√≥ √© conhecida ap√≥s o agrupamento ('GROUP BY'), essa condi√ß√£o deve aparecer em uma cl√°usula `HAVING`.

```python
select regi√£o, count(*), min(popula√ß√£o),
           max(popula√ß√£o), avg(popula√ß√£o), sum(popula√ß√£o) as tpop
           from estados group by regi√£o
       having count(*) > 5
           order by tpop desc
```

Veja o programa completo utilizando `HAVING`:

```python
# T√≠tulo: Listando apenas as regi√µes com mais de 5 estados
##############################################################################
import sqlite3

print("Regi√£o Estados Popula√ß√£o  M√≠nima    M√°xima      M√©dia    Total (soma)")
print("====== =======          ========= ========== ==========  ============")
with sqlite3.connect("brasil.db") as conex√£o:
    for regi√£o in conex√£o.execute("""
        select regi√£o, count(*), min(popula√ß√£o),
               max(popula√ß√£o), avg(popula√ß√£o), sum(popula√ß√£o) as tpop
        from estados
        group by regi√£o
        having count(*) > 5
        order by tpop desc"""):
        print("{0:6} {1:7} {2:18,} {3:10,} {4:10,.0f} {5:13,}".format(*regi√£o))
```
